```python
# Don't modify this cell.
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from pprint import pprint

# Import API key
import api_keys

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# Don't modify this cell
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

A lot of starter code has been generated for you. Use the cell below to play with the data to ensure you know what's happening. 

```python
# slice cities to see some randomly selected cities!
print(cities[0:20]) 
# see the accompanying latitudes and longitudes
print(lats[0:20])
print(lngs[0:20])
# when I Google the latitude and longitude of a given city in the list, it is pretty close to the values here
# this means that .nearest_city() worked!
```

### Perform API Calls
* Perform a weather check on each city in `cities` using a series of successive API calls.
* OPTIONAL: Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
# OpenWeatherMap API Key
api_key = api_keys.api_key

# base url for getting api data
base_url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + api_key

# example request
# req = requests.get(base_url + f'&q={cities[0]}').json()
import time

# create an empty list (append each dictionary to this)
city_weather = []

# create a for loop to iterate through each city in the list and create dictionaries with the data
for index, i in enumerate(cities):
    #print(index) 
    weather = requests.get(base_url + f'&q={cities[index]}').json()
    time.sleep(1)
    city_weather.append(weather)  

# Your code here. A loop maybe? 
#NOTE: API calls can be slow. They can also be limited. Once you have a list try not to rerun all API calls very often. 
# review the notes in Slack on how to pause - from Michelle and Matt
```

### Convert Raw Data to DataFrame
* OPTIONAL: Export the city data into a .csv.
* Display the DataFrame

```python
# see an example of a city's data
print(city_weather[0])

# create an empty list and populate it with cities that have data
clean_weather = []
for data in city_weather:
    if data['cod'] ==200:
        clean_weather.append(data)

# iterate over each city to collect data on different variables
lat = []
long = []
cloud = []
country = []
date = []
humid = []
max_temp = []
wind_speed = []

for index, city in enumerate(clean_weather):
    city = clean_weather[index]
    for key in city:
        if key == 'coord':
            latitude = city['coord']['lat']
            lat.append(latitude)
            longitude = city['coord']['lon']
            long.append(longitude)
        if key == 'clouds':
            cloudiness = city['clouds']['all']
            cloud.append(cloudiness)
        if key == 'sys':
            ctry = city['sys']['country']
            country.append(ctry)
        if key == 'dt':
            dt = city['dt']
            date.append(dt)
        if key == 'main':
            hum = city['main']['humidity']
            humid.append(hum)
            mx_temp = city['main']['temp_max']
            max_temp.append(mx_temp)
        if key == 'wind':
            wind_spd = city['wind']['speed']
            wind_speed.append(wind_spd)
            
# convert variable lists to Series  
lat_series = pd.Series(lat)
long_series = pd.Series(long)
cloud_series = pd.Series(cloud)
country_series = pd.Series(country)
date_series = pd.Series(date)
humid_series = pd.Series(humid)
max_temp_series = pd.Series(max_temp)
wind_speed_series = pd.Series(wind_speed)

# convert the list of dictionaries to a DataFrame and add columns/Series for each variable
weather_df = pd.DataFrame(clean_weather)
weather_df["Cloudiness"] = cloud_series
weather_df["Country"] = country_series
weather_df["Date"] = date_series
weather_df["Humidity"] = humid_series
weather_df["Lat"] = lat_series
weather_df["Lng"] = long_series
weather_df["Max Temp"] = max_temp_series
weather_df["Wind Speed"] = wind_speed_series

# drop unwanted columns and rename the City column
drop_weather_df = weather_df.drop(['coord', 'weather', 'base', 'main', 'visibility', 'wind', 'clouds','dt','sys','timezone','id','cod','rain'], axis=1)
wth_df = drop_weather_df.rename(columns={'name':'City'})

# review the final DataFrame
wth_df.head()

```
